```python
# Step 1: Normalize DRG and A_PDX
df_hist_sub_hum_ms_drg_age_los['ADRG'] =df_hist_sub_hum_ms_drg_age_los['ADRG'].astype(str).str.strip()
df_hist_sub_hum_ms_drg_age_los['PRIM_DX'] = df_hist_sub_hum_ms_drg_age_los['PRIM_DX'].astype(str).str.strip().str.upper()
df_hist_sub_hum_ms_drg_age_los['LOB'] = df_hist_sub_hum_ms_drg_age_los['LOB'].astype(str).str.strip().str.upper()
df_hist_sub_hum_ms_drg_age_los['PaymentType'] = df_hist_sub_hum_ms_drg_age_los['PaymentType'].str.strip().str.upper()

# Step 2: Ensure numeric fields are clean

df_hist_sub_hum_ms_drg_age_los['IDSavings'] = pd.to_numeric(df_hist_sub_hum_ms_drg_age_los['IDSavings'], errors='coerce').fillna(0)
df_hist_sub_hum_ms_drg_age_los['LOS'] = pd.to_numeric(df_hist_sub_hum_ms_drg_age_los['LOS'], errors='coerce').fillna(0)
df_hist_sub_hum_ms_drg_age_los['AGE'] = pd.to_numeric(df_hist_sub_hum_ms_drg_age_los['AGE'], errors='coerce').fillna(0)
df_hist_sub_hum_ms_drg_age_los['A_MCCCount'] = pd.to_numeric(df_hist_sub_hum_ms_drg_age_los['A_MCCCount'], errors='coerce').fillna(0).astype(int)
df_hist_sub_hum_ms_drg_age_los['A_CCCount'] = pd.to_numeric(df_hist_sub_hum_ms_drg_age_los['A_CCCount'], errors='coerce').fillna(0).astype(int)
```

```python
df1 = df_hist_sub_hum_ms_drg_age_los.copy()

adx_cols = [f"A_DX{i}" for i in range(2, 26)]
apx_cols = [f"A_PX{i}" for i in range(1, 26)]
bdx_cols = [f"B_DX{i}" for i in range(2, 26)]
bpx_cols = [f"B_PX{i}" for i in range(1, 26)]

df1["A_DX_List"] = df1[adx_cols].values.tolist()
df1["A_PX_List"] = df1[apx_cols].values.tolist()
df1["B_DX_List"] = df1[bdx_cols].values.tolist()
df1["B_PX_List"] = df1[bpx_cols].values.tolist()

df1["A_DX_List"] = df1["A_DX_List"].apply(lambda x: tuple(sorted([i for i in x if pd.notna(i)])))
df1["A_PX_List"] = df1["A_PX_List"].apply(lambda x: tuple(sorted([i for i in x if pd.notna(i)])))
df1["B_DX_List"] = df1["B_DX_List"].apply(lambda x: tuple(sorted([i for i in x if pd.notna(i)])))
df1["B_PX_List"] = df1["B_PX_List"].apply(lambda x: tuple(sorted([i for i in x if pd.notna(i)])))

df2 = df1.copy()

def clean_sdx_list(lst):
    cleaned = []
    for item in lst:
        item = str(item).upper().strip()               # normalize
        item = item.replace("- MCC", "")               # remove MCC
        item = item.replace("- CC", "")                # remove CC
        item = item.replace("MCC", "")                 # safety
        item = item.replace("CC", "")                  # safety
        cleaned.append(item.strip().replace("-", ""))  # final cleanup
    return tuple(cleaned)  # convert to tuple

df2["A_DX_List_Clean"] = df2["A_DX_List"].apply(clean_sdx_list)
df2["B_DX_List_Clean"] = df2["B_DX_List"].apply(clean_sdx_list)
```

```python
mccandcclist_df = pd.read_excel(r"C:\Users\Akumar\Downloads\MCCCCList.xlsx")
mccandcclist_df['ICDCode'] = (
    mccandcclist_df['ICDCode']
    .astype(str)
    .str.strip()
    .str.upper()
)
mccandcclist = dict(zip(mccandcclist_df['ICDCode'], mccandcclist_df['MCCorCC']))

def map_sdx_types(icd_list, lookup_dict):
    cleaned_list = []
    for code in icd_list:
        code_clean = str(code).strip().upper()  # normalize

        tag = lookup_dict.get(code_clean, "")  # MCC / CC / General
        cleaned_list.append(f"{code_clean} - {tag}")

    return tuple(cleaned_list)   # convert to tuple

df2['A_DX_Type_list'] = df2['A_DX_List_Clean'].apply(lambda x: map_sdx_types(x, mccandcclist))
df2['B_DX_Type_list'] = df2['B_DX_List_Clean'].apply(lambda x: map_sdx_types(x, mccandcclist))

```

```python
def split_px_list(px_list):
    mcc = []
    cc = []
    general = []
    
    for item in px_list:
        parts = item.split('-')
        code = parts[0].strip()
        flag = parts[1].strip() if len(parts) > 1 else ""
        
        if flag == "MCC":
            mcc.append(code)
        elif flag == "CC":
            cc.append(code)
        else:
            general.append(code)
    
    return tuple(mcc), tuple(cc), tuple(general)   # convert to tuple

df2["A_MCC_List"], df2["A_CC_List"], df2["A_General_List"] = zip(*df2["A_DX_Type_list"].apply(split_px_list))
df2["B_MCC_List"], df2["B_CC_List"], df2["B_General_List"] = zip(*df2["B_DX_Type_list"].apply(split_px_list))
```

```python
def extract_sdx_info(sdx_list):

    # If list/tuple is empty  return
    if not sdx_list:
        return "No SDX Present", tuple()

    # Separate entries by type
    mcc = [x.split(" - ")[0] for x in sdx_list if "- MCC" in x]
    cc  = [x.split(" - ")[0] for x in sdx_list if "- CC" in x]
    other = [
        x.split(" - ")[0]
        for x in sdx_list 
        if "- MCC" not in x and "- CC" not in x
    ]

    # Determine type and return tuple
    if mcc:
        return "MCC", tuple(mcc)
    elif cc:
        return "CC", tuple(cc)
    elif other:
        return "General", tuple(other)
    else:
        return "No SDX Present", tuple()

df2[["A_DX_Type", "A_DX_Set"]] = df2["A_DX_List"].apply(
    lambda x: pd.Series(extract_sdx_info(x))
)

df2[["B_DX_Type", "B_DX_Set"]] = df2["B_DX_List"].apply(
    lambda x: pd.Series(extract_sdx_info(x))
)
```
